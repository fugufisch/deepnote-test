```
import folium
from folium.plugins import HeatMapWithTime, HeatMap
```

Sadly the time lapse removes all plants w/o time information, but it's still cool to see.

```
decades = data.assign(decade=lambda x: (x.commissioned/10).round()*10)
points = (decades[['lat', 'lon', 'decade']]
    .dropna()
    .groupby('decade')
    .agg(list)
    .apply(lambda x: [[lat, lon] for lat, lon in zip(x.lat, x.lon)], axis=1)
    .to_list()
)
```

```
points = data[['lat', 'lon']].dropna().to_records(index=False)
points = [list(x) for x in points]
```

```
zoom_level = 5
```

```
m = folium.Map(location=[52.5243700,13.4105300], tiles="Stamen Toner", zoom_start=zoom_level)
hm = HeatMapWithTime(data=points)
m.add_child(hm)
```
